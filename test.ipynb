```python
"""假动态
"""
plt.figure()
plt.xlabel('x (Time)')
plt.ylabel('value')
plt.grid(True)

pid_time_array = [1,2,3,4,5]
pid_output_array = [21.00,30.00,50.00,100.00,255.00]
pid_input_array = [5.00,5.1,7.2,8.00,9.00]

max_output = 255

setpoint = 15
plt.ion()

for i in range(5):
    pid_time_array.append(pid_time_array[-1]+1)
    pid_output_array.append(i*2)
    pid_input_array.append(i*2+3)
    
    plt.clf()
    plt.grid(True)
    plt.xlabel('x (Time)')
    plt.ylabel('value')
    plt.plot(pid_time_array,pid_input_array,'-r',label="Temp")
    plt.plot(pid_time_array,[i *(max_output/255) for i in pid_output_array],'-b',label="pid output")
    plt.plot([pid_time_array[0],pid_time_array[-1]],[setpoint,setpoint],'-.y',label="setpoint")

    plt.legend(loc="best")
    plt.pause(1)
    
plt.ioff()
plt.show()
```
